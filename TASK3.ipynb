```python
import numpy as np
import pandas as pd
data=pd.read_csv(r"C:\Users\rajes\APEX PLANET INTERNSHIP PROJECT\final_cleaned_dataset.csv")
print(data)
```

```python
#create KPI Keys
total_emp=len(data)
print("Total_Employees:",total_emp)
```

```python
avg_salary=data["Salary"].mean()
print("Average Salary:",avg_salary)
```

```python
avg_age=data["Age"].mean()
print("Average Age:",avg_age)
```

```python
remote_percent=(data["Remote_Work"].value_counts(normalize=True).get(True,0)*100)
print("Remote_Work % :",remote_percent)
```

```python
#create Age Groups
data["Age Group"]=pd.cut(data["Age"],bins=[20,25,30,35,40,45],labels=["20-25","26-30","31-35","36-40","41-45"])
data.groupby("Age_Group")["Salary"].mean()
```
