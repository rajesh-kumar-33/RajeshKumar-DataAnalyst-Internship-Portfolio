```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv(r"C:\Users\rajes\APEX PLANET INTERNSHIP PROJECT\final_cleaned_dataset.csv")
print(data)
```

```python
# for stastical summary
data.describe()
```

```python
# for categorical summary
data.describe(include="object")
```

```python
# to create a histogram

data["Salary"].hist()
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
# histogram for age
data["Age"].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# to create barchat

data["Status"].value_counts().plot(kind="bar")
plt.title("Employee Status")
plt.show()
```
